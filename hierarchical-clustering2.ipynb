```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import matplotlib
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#Storing and reading the data.

fileName = "/kaggle/input/countries-continents/countries_continents.csv"

df = pd.read_csv(fileName)
df.head()
```

```python
df.dropna()

location = df[['Longitude','Latitude']]

location
```

**Q2.2**

```python
#Array of colors only for plotting the data
colors=['pink','lightseagreen','lightgreen','orange','lightcoral','lightblue']
```

1. **--SINGLE--**

```python
single_linkage = AgglomerativeClustering(n_clusters=7, linkage = 'single')
s_Linked = single_linkage.fit_predict(location)

df['linkage']= s_Linked
```

```python
#Plotting the data
plt.figure(figsize = (7,7))

#Scattter the data
plt.scatter(df['Longitude'], df['Latitude'], c = df['linkage'],cmap=matplotlib.colors.ListedColormap(colors),s=20)

plt.title('single linkage')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

2. **--COMPLETE--**

```python
#Aggolomerative object to apply complete method , i decided 7 clusters because of the reslutes of Lab2 
complete_linkage = AgglomerativeClustering(n_clusters=7, linkage = 'complete')
#Fit the model
comp_Linked = complete_linkage.fit_predict(location)

df['linkage']= comp_Linked
```

```python
#Plotting the data
plt.figure(figsize = (7,7))

#Scattter the data
plt.scatter(df['Longitude'], df['Latitude'], c = df['linkage'],cmap=matplotlib.colors.ListedColormap(colors),s=20)

plt.title('Complete linkage')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

3. **--AVERAGE--**

```python
#Aggolomerative object to apply average method , i decided 7 clusters because of the reslutes of Lab2 

average_linkage = AgglomerativeClustering(n_clusters=7, linkage = 'average')

#Fit the model
average = average_linkage.fit_predict(location)

df['linkage']= average
```

```python
#Plotting the data
plt.figure(figsize = (7,7))

#Scattter the data
plt.scatter(df['Longitude'], df['Latitude'], c = df['linkage'],cmap=matplotlib.colors.ListedColormap(colors),s=20)

plt.title('Average Linkage')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

4. **--WARD--**

```python
#Aggolomerative object to apply ward method , i decided 7 clusters because of the reslutes of Lab2 

ward_linkage = AgglomerativeClustering(n_clusters=7, linkage = 'ward')

#Fit the model
ward = ward_linkage.fit_predict(location)

df['linkage']= ward
```

```python
#Plotting the data
plt.figure(figsize = (7,7))

#Scattter the data
plt.scatter(df['Longitude'], df['Latitude'], c = df['linkage'],cmap=matplotlib.colors.ListedColormap(colors),s=20)

plt.title('Ward Linkage')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

**Q2.3: Dendogram**

```python
#Plotting the dendogram
sigle_linked = linkage(df[['Longitude','Latitude']], method = 'single')

plt.figure(1,figsize = (40,40))

dendrogram(sigle_linked,
           orientation='top',
           distance_sort='descending')

plt.title('Single Linkage',fontsize=30)
plt.xlabel('locations',fontsize = 30)
plt.ylabel('Euclidean distances',fontsize = 30)

plt.show()
```

**I chose to cut the dendogram at y= 30**

**Q2.5 Comparing with the actual clusters.**
